```python
from analysis.people import find_all_matches, con
import numpy as np
```

```python
SCORE_CUTOFF = 10.5
df_all = find_all_matches(con)
df = df_all[df_all["overall_score"] > SCORE_CUTOFF]
```

```python
filtered = df[df["krs_name"] == "Adam Borysiewicz"]
display(filtered)

def is_unique(s):
    a = s.to_numpy()
    return (a[0] == a).all()

non_unique_cols = []
for col in filtered.columns:
    if col.startswith("metaphone") or col.startswith("elections") or col.startswith("teryt"):
        continue
    if not is_unique(filtered[col]):
        non_unique_cols.append(col)
        
display(filtered[non_unique_cols])
```
